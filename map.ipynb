```python
import plotly.express as px

# Load built-in world map data
df = px.data.gapminder().query("year==2007")  # We just need country names & locations

# Create choropleth map
fig = px.choropleth(
    df,
    locations="iso_alpha",        # Country code (3-letter ISO code)
    color="continent",            # Color by continent
    hover_name="country",         # Show country name on hover
    projection="natural earth",   # Earth projection type
    title="üåç World Map - Hover to see Country"
)

# Show interactive map
fig.show()
```
